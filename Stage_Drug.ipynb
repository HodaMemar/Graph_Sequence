### Load Subject_id  as csv file

```python
import pandas as pd
df = pd.read_csv('D:\Embedding\Code\Data_Hemorhage\Subarachnoid.csv')
df.head()
```

```python
import mysql.connector

mydb = mysql.connector.connect(
  host="localhost",
  user="root",
  password="azkaban123456",
  database="mimiciiiv14"
)
```

```python
txt_counter=0
Drug_Metadata=[]
mycursor = mydb.cursor()
for i in range(len(df)):
    
    v_SUBJECT_ID=str(df.iloc[i].SUBJECT_ID)
    v_HADM_ID=str(df.iloc[i].HADM_ID)
    str_Drug="SELECT SUBJECT_ID,HADM_ID,STARTDATE,ENDDATE,DRUG FROM mimiciiiv14.prescriptions where SUBJECT_ID="+v_SUBJECT_ID+ " and HADM_ID="+v_HADM_ID+" and DRUG_TYPE='MAIN'  and STARTDATE is not null and ENDDATE  is not null "
    
    mycursor.execute(str_Drug)
    #num_fields = len(mycursor.description)
    #field_names = [i[0] for i in mycursor.description]
    #print(v_SUBJECT_ID,field_names)
    myresult = mycursor.fetchall()
    for ele in myresult:
        res = list(ele)  
        Drug_Metadata.append(res)
        
df_drug=pd.DataFrame(Drug_Metadata,columns=['SUBJECT_ID','HADM_ID','STARTDATE','ENDDATE','DRUG']) 

```

### Expand Date by day

```python
from datetime import date, timedelta
Stage_ls_Drug=[]
for c in range(len(df_drug)):
    sdate=df_drug.iloc[c]['STARTDATE']
    edate=df_drug.iloc[c]['ENDDATE']
    v_SUBJECT_ID=df_drug.iloc[c]['SUBJECT_ID']
    v_HADM_ID=df_drug.iloc[c]['HADM_ID']
    v_DRUG=df_drug.iloc[c]['DRUG']
    
    delta = edate - sdate       # as timedelta
    print(v_SUBJECT_ID,delta)
    i=0
    for i in range(delta.days + 1):
        day = sdate + timedelta(days=i)
        Stage_ls_Drug.append([v_SUBJECT_ID,1,v_HADM_ID,0,str(day.date()),'Drug','Main',v_DRUG])

Stage_df_Drug=pd.DataFrame(Stage_ls_Drug,columns=['Subject_id','Subgraph_id','admission_Id','ago','Timestame_id','Event type','entity','value']) 
Stage_df_Drug
```

```python
Stage_df_Drug.to_csv('D:\Embedding\Code\Data_Hemorhage\Stage_Drug_Subarachnoid.csv', index = False)
```
