# Merge Data sets

```python
frames = [Subarachnoid,Intracerebral]
repository_df = pd.concat(frames)
repository_df=repository_df.reset_index(drop=False)
repository_df.shape
```

```python
#repository_df=Malignant
```

```python
from sklearn.utils import shuffle
repository_df = shuffle(repository_df)
repository_df
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
```

```python
labels = repository_df['Target']
features =  repository_df.iloc[:, 1:21]
features
```

```python
print(labels.shape)
print(features.shape)
```

# train_test_split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size = 0.2)
```

# SVM - Binary

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
